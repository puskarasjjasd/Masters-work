```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(123)
simulation_num = 200

all_walks=[]

for i in range(simulation_num):
    random_walk = [0]
    
    for x in range(100):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
    
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        
        random_walk.append(step)
        
    plt.plot(random_walk)
    all_walks.append(random_walk[-1])
    
    
plt.show()

print('The final steps achieved in each simulation: \n' , all_walks)

plt.hist (all_walks)
plt.xlabel('final steps achieved')
plt.ylabel('frequency')
plt.show()

all_walks = np.array(all_walks)

count=len(all_walks[all_walks >60])
print("The probability of step account >= 60 is: ", count/simulation_num)
```
